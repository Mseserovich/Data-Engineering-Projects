#### Fix Column names

```python
def fix_column_names(columns):
    pattern = r"(\[)(?P<columns>.+?)(\])"
    new_column = re.search(pattern, columns)
    if new_column:
        return new_column.group("columns")
```

```python
new_columns = [*map(fix_column_names, df_person_scores.columns)]
```

#### Convert values inside tuples to integers

```python
def convert_to_integers(df):
    array = np.array(df)
    for i, j in enumerate(array):
        p = list(j)
        for k, l in enumerate(p):
            nums = re.findall(r"\d", l)
            nums2 = [*map(int, nums)]
            array[i][k] = tuple([int(i) for i in nums2])
    return array
```

```python
def sum_scores(tuples):
    scale, score = tuples
    if score == 5:
        sub_scales[str(scale)].append(1)
```

```python
DICT = {
    "conscientiousness" : [],
    "agreeableness" : [],
    "emotional_stability" : [],
    "openness_to_new_experience" : [],
    "extraversion" : [],
    
}

sub_scales = {
    "1" : [],
    "2" : [],
    "3" : [],
    "4" : [],
    "5" : []
}



def calculate_subscale():
    for i in range(1555):
        df_merged.iloc[i, 0:50].apply(sum_scores)
        DICT["extraversion"].append(sum(sub_scales["1"]))
        sub_scales['1'].clear()
        DICT["agreeableness"].append(sum(sub_scales["2"]))
        sub_scales['2'].clear()
        DICT["openness_to_new_experience"].append(sum(sub_scales["5"]))
        sub_scales['5'].clear()
        DICT["emotional_stability"].append(sum(sub_scales["4"]))
        sub_scales['4'].clear()
        DICT["conscientiousness"].append(sum(sub_scales["3"]))
        sub_scales['3'].clear()
```

```python
calculate_subscale()
```
